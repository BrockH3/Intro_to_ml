```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics
from google.colab import drive
drive.mount('/content/drive')

filepath='/content/drive/My Drive/Intro to ML/datasets/diabetes.csv'
diabetes = pd.DataFrame(pd.read_csv(filepath))
diabetes.head()
```

```python
#problem 1
#split the training and validation sets
from sklearn.model_selection import train_test_split
np.random.seed(0)
train, val = train_test_split(diabetes, train_size = .8, test_size = .2, random_state = 100)

#set inputs and outputs
y_train = train.pop('Outcome')
x_train = train

y_val = val.pop('Outcome')
x_val = val

#scale input between 1 and 0
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_val = sc_X.transform(x_val)
```

```python
#linear classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)

y_pred = classifier.predict(x_val)
```

```python
#display accuracy, precision, and recall
print("Accuracy:",metrics.accuracy_score(y_val, y_pred))
print("Precision:",metrics.precision_score(y_val, y_pred))
print("Recall:",metrics.recall_score(y_val, y_pred))
print("F1:",metrics.f1_score(y_val, y_pred))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_val, y_pred)
print(confusion_matrix)
```

```python
# problem 2
filepath2='/content/drive/My Drive/Intro to ML/datasets/cancer.csv'
cancer = pd.DataFrame(pd.read_csv(filepath2))

#drop id column and turn results from M and B to 1 and 0
cancer = cancer.drop(columns = ['id','Unnamed: 32'])

cancer.head()
```

```python

def binary_map(x):
    return x.map({'M': 1, 'B': 0,})

cancer[['diagnosis']] = cancer[['diagnosis']].apply(binary_map)

cancer.head()
```

```python

np.random.seed(0)
train, val = train_test_split(cancer, train_size = .8, test_size = .2, random_state = 100)

#set inputs and outputs
y_train = train.pop('diagnosis')
x_train = train

y_val = val.pop('diagnosis')
x_val = val

#scale input between 1 and 0
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_val = sc_X.transform(x_val)
```

```python
#linear classifier code
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)

y_pred = classifier.predict(x_val)
```

```python
#print accuracy, precision, and recall
print("Accuracy:",metrics.accuracy_score(y_val, y_pred))
print("Precision:",metrics.precision_score(y_val, y_pred))
print("Recall:",metrics.recall_score(y_val, y_pred))
print("F1:",metrics.f1_score(y_val, y_pred))
```

```python
#print confusion matric
from sklearn.metrics import confusion_matrix
confusion_matrix2 = confusion_matrix(y_val, y_pred)
print(confusion_matrix2)
```

```python
#added weight penalty
classifier_weight = LogisticRegression(random_state=0, penalty = 'l1', solver='liblinear')
classifier_weight.fit(x_train,y_train)

y_pred = classifier.predict(x_val)

#print results
print("Accuracy:",metrics.accuracy_score(y_val, y_pred))
print("Precision:",metrics.precision_score(y_val, y_pred))
print("Recall:",metrics.recall_score(y_val, y_pred))
print("F1:",metrics.f1_score(y_val, y_pred))
```

```python
#naive bayesian model
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train,y_train)
y_pred = nb.predict(x_val)

print(metrics.classification_report(y_val, y_pred))
```

```python
#seperate diagnosis from features for scaling
cancer2=cancer
y=cancer2.pop('diagnosis')
x=cancer2
```

```python
#PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents)
principalDf.head()
```

```python
#scale PCA
principalDF_scale = sc_X.fit_transform(principalDf)

#add target back to set
DF_final = pd.concat([pd.DataFrame(principalDF_scale), y], axis=1)
DF_final.head()
```

```python
#seperate train from test sets
np.random.seed(0)
train, val = train_test_split(DF_final, train_size = .8, test_size = .2, random_state = 100)

y_train = train.pop('diagnosis')
x_train = train

y_val = val.pop('diagnosis')
x_val = val

#linear regression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)

y_pred = classifier.predict(x_val)
#print results
print("Accuracy:",metrics.accuracy_score(y_val, y_pred))
print("Precision:",metrics.precision_score(y_val, y_pred))
print("Recall:",metrics.recall_score(y_val, y_pred))
print("F1:",metrics.f1_score(y_val, y_pred))
```
