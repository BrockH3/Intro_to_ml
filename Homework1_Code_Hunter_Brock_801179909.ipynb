```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python

url='/content/drive/My Drive/Intro to ML/datasets/D3.csv'
df=pd.read_csv(url)     #import csv file to readable format in python

x1=df.values[:,0]       #input first column
x2=df.values[:,1]       #input second column
x3=df.values[:,2]       #input third column
y=df.values[:,3]        #input fourth column

#length for number of training samples
m=len(y)
n=len(x1)

#create feature matrix for operation
X_0 = np.ones((m,1))

X_1 = x1.reshape(m,1)
X_2 = x2.reshape(m,1)
X_3 = x3.reshape(m,1)

X1 = np.hstack((X_0,X_1))
X2 = np.hstack((X_0,X_2))
X3 = np.hstack((X_0,X_3))

theta = np.zeros(2)
```

```python
#computes the loss of linear regression
def loss_function(x,y,theta):

  prediction = x.dot(theta)
  error = np.subtract(prediction,y)
  error_square = np.square(error)

  Out = 1/(2*m) * np.sum(error_square)

  return Out
```

```python
cost1 = loss_function(X1,y,theta)
print('cost of x1 = ',cost1)
cost2 = loss_function(X2,y,theta)
print('cost of x2 = ',cost2)
cost3 = loss_function(X3,y,theta)
print(' cost of x3 = ', cost3)
```

```python
def gradient_descent(x,y,theta,alpha,iterations):
  #compute the gradient descent for the linear regression

  #alpha is learning rate

  m=len(y)
  cost_history = np.zeros(iterations)     #creates a matrix to store the history

  for i in range(iterations):
    predictions = x.dot(theta)            #x dot product theta
    errors = np.subtract(predictions, y)  #calculate error

    #calculates the sum function for gradient descent
    sum_delta = (alpha / m) * x.transpose().dot(errors)
    theta -= sum_delta
    cost_history[i] = loss_function(x, y, theta)

  return theta,cost_history
```

```python
theta=[0.,0.]
iterations = 1000
alpha = 0.05
```

```python
theta,cost_history = gradient_descent(X1,y,theta,alpha,iterations)
```

```python
#scatter plot and linear regression plot for x1
plt.scatter(X1[:,1],y,color='red',marker='+',label ='X1 Data')
plt.plot(X1[:,1], X1.dot(theta), color='green', label='Linear Regression')

plt.grid(True)
plt.xlabel('X1')
plt.ylabel('y')
plt.title('Linear Regression Model for X1')
plt.show()
```

```python
plt.plot(range(1, iterations + 1), cost_history, color='blue')
plt.grid(True)

plt.xlabel('Iterations')
plt.ylabel('Cost')

plt.show()
```
